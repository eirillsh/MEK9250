```python
import matplotlib.pyplot as plt
import dolfinx as dfx
import numpy as np
import sympy as sp
```

```python
x=np.linspace(0,1,101)
u_e = -1/12*(x**4-x)
u=np.zeros(10)

for k in range(1,11):
    if k%2==0:
        u[k-1]=-2/(np.pi*k)**3
    else:
        u[k-1]=2/(np.pi*k)**3-8/(np.pi*k)**5

u_fun=0
for i in range(0,10):
    u_fun+=u[i]*np.sin(np.pi*(i+1)*x)

plt.plot(x,u_fun,'r',label = 'FEM')
plt.plot(x,u_e,'b--',label="Exact")
plt.legend()
plt.show()
```

```python
plt.plot(x,u_e-u_fun,'r',label = 'error')
plt.legend()

f=(u_e-u_fun)**2
I_trap = (1/200)*(f[0] + \
          2 * sum(f[1:100]) + f[100])

print(f"L_2 error is : {np.sqrt(I_trap):.2e}")
print(f"L_inf error is : {np.max(abs(u_e-u_fun)):.2e}")
```
